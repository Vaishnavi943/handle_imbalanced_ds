# Imbalanced dataset

A dataset with an unequal class distribution.

```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv('credit_data.csv')
```

```python
dataset.head()
```

```python
# 2 class present
# legit transaction(normal) = 0
# fraud transaction = 1
```

```python
dataset.shape
```

```python
# distribution of 2 classes
dataset['Class'].value_counts()
```

```python
# separating the legit and fraud transaction
legit = dataset[dataset['Class'] == 0]
fraud = dataset[dataset['Class'] == 1]
```

```python
legit.shape, fraud.shape
```

```python
# Under-sampling -> build a sample dataset containing similar distribution of legit and fraud transaction
legit_sample = legit.sample(n=3)
legit_sample.shape
```

```python
# concate the 2 dataframes
new_dataset = pd.concat([legit_sample, fraud], axis = 0)  # axis = 0 --> concat row-wise
```
